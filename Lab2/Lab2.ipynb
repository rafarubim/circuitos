$V_{N^{-}} = V_{N^{+}} = 0$

A corrente em $R_{1}$ deve ser:

$i_{R_{1}} = (V_{in} - V_{N^{-}}) / R_{1} = V_{in} / R_{1}$

&emsp;Essa corrente deve corresponder à corrente em $R_{2}$, pois idealmente não há corrente na entrada inversora do AmpOp. Temos:

$i_{R_{2}} = (V_{N^{-}} - V_{out}) / R_{2} = i_{R_{1}}$

&emsp;Portanto:

$-V_{out} / R_{2} = V_{in} / R_{1}$

$V_{out} = - V_{in} \cdot R_{2} / R_{1}$

&emsp;Estendendo o resultado para a saturação, tem-se:

- No limite $V^{+}$:

$V_{out} \geq V^{+} \Rightarrow - V_{in} \cdot R_{2} / R_{1} \geq V^{+}$

$\Rightarrow V_{in} \leq - V^{+} \cdot R_{1} / R_{2}$

- No limite $V^{-}$:

$V_{out} \leq V^{-} \Rightarrow - V_{in} \cdot R_{2} / R_{1} \leq V^{-}$

$\Rightarrow V_{in} \geq - V^{-} \cdot R_{1} / R_{2}$

&emsp;Logo, obteve-se os limites de saturação:

&emsp;Logo, obteve-se os limites de saturação:
