### a)

&emsp;O polinômio característico do circuito RLC é:

$p(\lambda) = \lambda^{2}+\dfrac{R}{L}.\lambda+\dfrac{1}{L.C}$

&emsp;Portanto as raízes(pólos) são:

$\lambda_{1} = -\dfrac{R}{2.L} + \dfrac{1}{2}.\sqrt{ \left(\dfrac{R}{L}\right)^{2}-\dfrac{4}{L.C}}$

$\lambda_{2} = -\dfrac{R}{2.L} - \dfrac{1}{2}.\sqrt{ \left(\dfrac{R}{L}\right)^{2}-\dfrac{4}{L.C}}$

&emsp;Logo a solução homogênea depende dos valores $R$, $L$ e $C$:

$S_{H}(t) =$

- Caso superamortecido

$\Large k_{1}.e^{\lambda_{1}.t}+k_{2}.e^{\lambda_{2}.t}$, se $\left(\dfrac{R}{L}\right)^{2}-\dfrac{4}{L.C} > 0$

- Caso criticamente amortecido

$\Large k_{1}.e^{\lambda.t}+k_{2}.t.e^{\lambda.t}$, se $\left(\dfrac{R}{L}\right)^{2}-\dfrac{4}{L.C} = 0$

- Caso subamortecido

$\Large k_{1}.cos(\beta.t).e^{\alpha.t}+k_{2}.sin(\beta.t).e^{\alpha.t}$, se $\left(\dfrac{R}{L}\right)^{2}-\dfrac{4}{L.C} < 0$ e $\lambda_{1} = \alpha + j.\beta$

&emsp;Assumindo que o circuito atinge o regime permanente em um semiperíodo da onda, sabe-se que quando a tensão vai de $A$ a $0$ pode-se aplicar a solução homogênea para calcular cada tensão no circuito. Vamos deslocar a onda para que isso ocorra em $t = 0$ para usarmos a solução homogênea como está e facilitar as contas. As tensões nos elementos quando a tensão da fonte é $A$ no regime permanente são:

$V_{R} = 0$, pois o capacitor é como um aberto e não há corrente no circuito

$V_{L} = 0$, pois o indutor é como um curto

$V_{C} = A$, pelo KVL na malha única

&emsp;Imediatamente após a onda entrar em seu primeiro semiperíodo (quando a tensão vai de $A$ a $0$, em $t = 0$), as condições iniciais das tensões nos elementos serão:

$V_{R} = 0$, pois a corrente no indutor é contínua e continua valendo $0$

$V_{C} = A$, pois a tensão no capacitor é contínua

$V_{L} = -A$, pelo KVL na malha única

$\dfrac{d.V_{C}}{dt} = 0$, pois $i = C.\dfrac{d.V_{C}}{dt}$ e a corrente vale $0$

$\dfrac{d.i}{dt} = -\dfrac{A}{L}$, pois $V_{L} = L.\dfrac{d.i_{L}}{dt}$ e a tensão no indutor vale $-A$

$\dfrac{d.V_{R}}{dt} = -\dfrac{R.A}{L}$, pois $\dfrac{d.V_{R}}{dt} = R.\dfrac{d.i}{dt}$ e a derivada da corrente vale $-\dfrac{A}{L}$

$\dfrac{d.V_{L}}{dt} = \dfrac{R.A}{L}$, pela derivada da KVL na malha única

&emsp;Substituindo as condições iniciais nas equações homogêneas de cada caso para cada elemento, obtém-se o valor das constantes $k_{1}$ e $k_{2}$ que determinam a expressão da tensão naquele elemento para $0 \leq t \leq \dfrac{T}{2}$. Em $t = \dfrac{T}{2}$ a tensão da fonte faz o contrário, mudando de $0$ a $A$. Esse caso exige uma solução particular $S_{P}(t)$ (onde as condições iniciais são nulas e há fontes atuantes a partir de $t = t_{0}$), mas essa solução será o inverso (negativo) da homogênea, pois a onda atinge o regime permanente a cada semiperíodo. Também é deslocada no tempo de maneira a começar em $t = \dfrac{T}{2}$ e tem novas condições iniciais que só farão diferença no caso do capacitor, que deve somar uma constante de modo que $S_{P}\left(\dfrac{T}{2}\right) = 0$ (pois no caso da solução particular o capacitor começa descarregado, ao invés de começar com tensão $A$ como na homogênea. Nos casos do indutor e resistor a condição inicial é o negativo do caso homogêneo).

&emsp;Assim teremos (as seguintes respostas só são válidas em um período ($0 \leq t \leq T$), mas se repetem para outros períodos da onda quadrada):

$\Large S(t) = S_{H}(t) + S_{P}(t)$

### Caso superamortecido:

- Capacitor:

$\Large V_{C}(t) = k_{1}.e^{\lambda_{1}.t}+k_{2}.e^{\lambda_{2}.t} + (k_{1} + k_{2}) - k_{1}.e^{\lambda_{1}.(t-\frac{T}{2})} - k_{2}.e^{\lambda_{2}.(t-\frac{T}{2})}$

&emsp;Tal que $k_{1}$ e $k_{2}$ são resultados do sistema:

$k_{1} + k_{2} = A$

$\lambda_{1}.k_{1} + \lambda_{2}.k_{2} = 0$

&emsp;Tal que $k_{1}$ e $k_{2}$ são resultados do sistema:

&emsp;Tal que $k_{1}$ e $k_{2}$ são resultados do sistema:

### b)

&emsp;O circuito de segunda ordem poderá ter até duas constantes de tempo.

&emsp;No caso superamortecido, tem-se $\tau_{1} = \dfrac{1}{|\lambda_{1}|}$ e $\tau_{2} = \dfrac{1}{|\lambda_{2}|}$. O tempo de assentamento deve ser calculado através da maior delas, que é obtida pelo pólo/raiz dominante (menor pólo/raiz em módulo). Como $min(\lambda_{1}, \lambda_{2}) = \lambda_{1}$ (trivialmente pelo valor de $\lambda_{1}$ e $\lambda_{2}$ dados na letra $(a)$), então a maior constante de tempo é $\tau_{1}$.

&emsp;No caso criticamente amortecido, só temos uma raiz, então temos a constante de tempo é $\tau = \dfrac{1}{|\lambda|}$.

&emsp;No caso subamortecido, podemos ignorar os senos e cossenos que oscilam e considerar apenas a exponencial que diminui o efeito da resposta transitória para encontrar o tempo de assentamento. Considera-se a constante de tempo nesse caso então $\tau = \dfrac{1}{|\alpha|}$.

&emsp;Conclui-se que o tempo $\dfrac{T_{min}}{2}$ (meia onda) é para cada caso:

- Caso superamortecido:

$\dfrac{T_{min}}{2} = 5.\tau_{1} = \dfrac{5}{|\lambda_{1}|}$

- Caso criticamente amortecido:

$\dfrac{T_{min}}{2} = 5.\tau = \dfrac{5}{|\lambda|}$

- Caso subamortecido:

$\dfrac{T_{min}}{2} = 5.\tau = \dfrac{5}{|\alpha|}$
