### Demonstração do funcionamento:

&emsp;Sejam $v^{+}$ e $v^{-}$ as tensões nas entradas não-inversora e inversora do AmpOp, respectivamente. As correntes em $R_{1}$ e $R_{2}$ são:

$i_{R_{1}} = (IN1 - v^{+}) / R_{1}$

$i_{R_{2}} = (v^{+} - IN2) / R_{2}$

&emsp;Idealmente não há corrente na entrada não-inversora, portanto $R_{1}$ e $R_{2}$ estão em série. Temos então:

$i_{R_{1}} = i_{R_{2}}$

$(IN1 - v^{+}) / R_{1} = (v^{+} - IN2) / R_{2}$

$v^{+} = (IN2 - v^{+}) \cdot R_{1} / R_{2} + IN1$

$v^{+} \cdot (R_{2} + R_{1}) / R_{2} = IN2 \cdot R_{1} / R_{2} + IN1$

$v^{+} = IN2 \cdot R_{1} / (R_{1} + R_{2}) + IN1 \cdot R_{2} / (R_{1} + R_{2})$

&emsp;Por realimentação, $v^{-} = v^{+}$. Idealmente não há corrente na entrada inversora, portanto $R_{5}$ e $R_{f2}$ estão em série. Logo:

$i_{R_{5}} = (0 - v^{-}) / R_{5}$

$i_{R_{f2}} = (v^{-} - V_{out}) / R_{f2}$

$i_{R_{5}} = i_{R_{f2}}$

$(0 - v^{-}) / R_{5} = (v^{-} - V_{out}) / R_{f2}$

$V_{out} = v^{-} (1 + R_{f2} / R_{5})$

$V_{out} = [IN2 \cdot R_{1} / (R_{1} + R_{2}) + IN1 \cdot R_{2} / (R_{1} + R_{2})] (R_{5} + R_{f2}) / R_{5}$

$V_{out} = IN2 \cdot R_{1} \cdot (R_{5} + R_{f2}) / [R_{5} \cdot (R_{1} + R_{2})] + IN1 \cdot R_{2} \cdot (R_{5} + R_{f2})/ [R_{5} \cdot (R_{1} + R_{2})]$

&emsp;Considerando $V_{out} = A \cdot IN1 + B \cdot IN2$, tem-se pela fórmula valores positivos de $A$ e $B$:

$A = R_{2} \cdot (R_{5} + R_{f2})/ [R_{5} \cdot (R_{1} + R_{2})]$

$B = R_{1} \cdot (R_{5} + R_{f2}) / [R_{5} \cdot (R_{1} + R_{2})]$
